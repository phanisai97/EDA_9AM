# Task -1
- I want to seperate the categorical(object is the dtype) and numerical columns(int64 or float is dtype)
- Catgorical columns in one list and numerical columns in another list

```python
cat_col=[]
num_col=[]
for i,j in zip(cols,dtype):
    if j=='object':
        cat_col.append(visa_df[i])
    else:
        num_col.append(visa_df[i])
print(cat_col)
print(num_col) 
```

```python
cat=[j for i,j in zip(cols,dtype) if j=='object']
num=[j for i,j in zip(cols,dtype) if j!='object']
len(cat),len(num)
```

# Shortcut is select dtypes


Signature: visa_df.select_dtypes(include=None, exclude=None) -> 'Self'
Docstring:
Return a subset of the DataFrame's columns based on the column dtypes.

Parameters
----------
include, exclude : scalar or list-like
    A selection of dtypes or strings to be included/excluded. At least
    one of these parameters must be supplied.

Returns
-------
DataFrame
    The subset of the frame including the dtypes in ``include`` and
    excluding the dtypes in ``exclude``.
