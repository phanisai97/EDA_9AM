```python
"""
The box extends from the first quartile (Q1) to the third
quartile (Q3) of the data, with a line at the median.
The whiskers extend from the box to the farthest data point
lying within 1.5x the inter-quartile range (IQR) from the box.
Flier points are those past the end of the whiskers.
See https://en.wikipedia.org/wiki/Box_plot for reference.

.. code-block:: none

          Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                       |-----:-----|
       o      |--------|     :     |--------|    o  o
                       |-----:-----|
     flier             <----------->            fliers
                            IQR
                            """
```

```python
# step1,2,3: Find q1,q2,q3
pwage=visa_df['prevailing_wage']
q1=np.percentile(pwage,25)
q2=np.percentile(pwage,50)
q3=np.percentile(pwage,75)
q1,q2,q3
# Step4: IQR=q3-q1
iqr=q3-q1
# Step 5,6 Find lb and ub
lb=q1-1.5*iqr
ub=q3+1.5*iqr
# Step7,8 Find con1 and con2
con1=wage_data>=lb
con2=wage_data<=ub
# Step9,10 Find con and count satisfying con
con= con1 & con2
count=len(visa_df[con])
count
```

# Method 1 of dealing with outliers
- I want to call the each observation of wage_data
- If that observation will available in this 427 values, then that observation will be replaced by median
- Otherwise, that observation is available as it is

```python
# I want to call the each observation of wage_data
# If that observation will available in this 427 values, then that observation will be replaced by median
# Otherwise, that observation is available as it is
pwage=visa_df['prevailing_wage']
outliers=outliers_df['prevailing_wage'].values.tolist()
pwage_median=pwage.median()
new_pwage=[]
for i in pwage:
    if i in outliers:
        new_pwage.append(pwage_median)
    else:
        new_pwage.append(i)
print(pwage_median,new_pwage)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.hist(pwage)
plt.subplot(2,2,2)
plt.hist(new_pwage)
plt.subplot(2,2,3)
plt.boxplot(pwage,vert=False)
plt.subplot(2,2,4)
plt.boxplot(new_pwage,vert=False)
plt.show()
```

# Windsorization,Method 2 of dealing with outliers
- whatever the outliers less than lb fill with lb
- whatever the outliers more than ub fill with ub
- in numpy, we have method **clip**

```python
# Step 1: get the wage data values
pwage=visa_df['prevailing_wage']
# Step 2 calc lb
# Step 3 calc ub
q1=np.percentile(pwage,25)
q2=np.percentile(pwage,50)
q3=np.percentile(pwage,75)
q1,q2,q3
# Step 4: IQR=q3-q1
iqr=q3-q1
# Step 5,6 Find lb and ub
lb=q1-1.5*iqr
ub=q3+1.5*iqr

# Step 4 check the parameters and apply in numpy np.clip()
pwage_clip=np.clip(pwage,lb,ub)
# step 5: again implement above visualization analysis
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.hist(pwage)
plt.subplot(2,2,2)
plt.hist(pwage_clip)
plt.subplot(2,2,3)
plt.boxplot(pwage,vert=False)
plt.subplot(2,2,4)
plt.boxplot(pwage_clip,vert=False)
plt.show()
```
